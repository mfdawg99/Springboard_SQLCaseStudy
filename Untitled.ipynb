Q3: Write an SQL query to show a list of facilities that charge a fee to members, where the fee is less than 20% of the facility's monthly maintenance cost. Return the facid, facility name, member cost, and monthly maintenance of the facilities in question.

```python
query = '''SELECT facid,name,membercost,monthlymaintenance           
            FROM Facilities 
            WHERE membercost <= (0.2 * monthlymaintenance);'''
df = pd.read_sql_query(query, engine)
df
```

Q4: Write an SQL query to retrieve the details of facilities with ID 1 and 5. Try writing the query without using the OR operator.

Q5: Produce a list of facilities, with each labelled as 'cheap' or 'expensive', depending on if their monthly maintenance cost is
more than $100. Return the name and monthly maintenance of the facilities in question.

```python
query = '''SELECT name,
                    monthlymaintenance,
            CASE WHEN monthlymaintenance <= 100 THEN 'cheap' 
                 WHEN monthlymaintenance > 100 THEN 'expensive'
                 ELSE '' END AS state
            FROM Facilities;'''
df = pd.read_sql_query(query, engine)
df
```

Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name. */


```python
#Bookings,Facilities,Members

query = ''' WITH history AS (SELECT *
                             FROM 
                                 Facilities AS f
                             LEFT JOIN 
                                 Bookings as b
                             ON b.facid = f.facid
                             )
                                
            SELECT DISTINCT
                    m.surname || ', ' || m.firstname as fullname,
                    history.name as facility
            FROM 
                Members as m
            LEFT JOIN 
                history 
            ON 
                history.memid = m.memid
            WHERE
                history.name LIKE 'Tennis Court%' AND
                fullname NOT LIKE '%GUEST%'
            ORDER BY
                fullname
            ;
            '''
df = pd.read_sql_query(query, engine)
df
```
