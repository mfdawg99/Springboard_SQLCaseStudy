```python
query = '''SELECT
              name,
              SUM(membercost + guestcost) * slots AS revenue
            FROM
                Facilities
            LEFT JOIN
                Bookings as book
            ON
                book.facid = Facilities.facid
            GROUP BY
                name
            ORDER BY
                revenue DESC
        ;'''
df = pd.read_sql_query(query, engine)
df
```

```python
query = '''SELECT
              name,
              SUM((membercost * slots) + (guestcost * slots)) AS revenue
            FROM
                Facilities
            LEFT JOIN
                Bookings as book
            ON
                book.facid = Facilities.facid
            GROUP BY
                name
            ORDER BY
                revenue DESC
        ;'''
df = pd.read_sql_query(query, engine)
df
```

```python
query = '''SELECT DISTINCT
              surname,
              firstname, 
              recommendedby
            FROM
                Facilities
            LEFT JOIN
                Bookings as book
                ON
                    book.facid = Facilities.facid
            LEFT JOIN
                Members as mem
                ON
                    mem.memid = book.memid
            WHERE 
                    recommendedby NOT LIKE ("")
                AND
                    surname NOT LIKE('%GUEST')
            ORDER BY
                recommendedby DESC
        ;'''
df = pd.read_sql_query(query, engine)
df
```

```python
query = '''SELECT 
             *
            FROM
                Facilities
            LEFT JOIN
                Bookings as book
                ON
                    book.facid = Facilities.facid
            LEFT JOIN
                Members as mem
                ON
                    mem.memid = book.memid
            WHERE
                slots = 1
                
        ;'''

df = pd.read_sql_query(query, engine)
df
```

```python
query = '''SELECT 
             count(slots) AS totalslots,
             surname || ', ' || firstname as membername
            FROM
                Facilities
            LEFT JOIN
                Bookings as book
                ON
                    book.facid = Facilities.facid
            LEFT JOIN
                Members as mem
                ON
                    mem.memid = book.memid
            WHERE
                    membername NOT LIKE ("%GUEST%")
                AND 
                    mem.memid != 0
            GROUP BY
                membername
            ORDER BY
                totalslots DESC
        ;'''

df = pd.read_sql_query(query, engine)
df
```

```python
query = '''SELECT 
                SUBSTR(starttime,6,2) AS month,
                name,
                SUM(slots) AS totalslots
            FROM
                Facilities
            LEFT JOIN
                Bookings as book
                ON
                    book.facid = Facilities.facid
            LEFT JOIN
                Members as mem
                ON
                    mem.memid = book.memid
            WHERE
                    surname NOT LIKE ("%GUEST%")
                AND 
                    mem.memid != 0
            GROUP BY
                name, month
            ORDER BY
                name
        ;'''

df = pd.read_sql_query(query, engine)
df
```
